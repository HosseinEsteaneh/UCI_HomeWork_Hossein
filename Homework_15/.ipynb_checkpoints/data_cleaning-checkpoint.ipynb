```python
import pandas as pd
```

```python
# csv file variable assignments

monthes = 12

for month in range(1,monthes+1):
    if month < 10:
        #print("path_"+str(month))
        print(f"data/JC-20200{month}-citibike-tripdata.csv")
    else:
        print(f"data/JC-2020{month}-citibike-tripdata.csv")
    
```

```python
# Appending all CSV files in the folder into one dataframe

# csv file variable assignments

monthes = 12
total_df=[]

for month in range(1,monthes+1):
    #path = "month_"+month
    if month == 1:
        path = str("data/JC-20200"+str(month)+"-citibike-tripdata.csv")
        total_df = pd.read_csv(path)
        print(path)
        print(total_df.value_counts().count())

    if (month < 10)&(month>1):
        path = str("data/JC-20200"+str(month)+"-citibike-tripdata.csv")
        month_df = pd.read_csv(path)
        print(path)
        total_df = total_df.append(month_df, ignore_index=True)
        print(total_df.value_counts().count())

    if month >=10:
        path = "data/JC-2020"+str(month)+"-citibike-tripdata.csv"
        print(path)
        month_df = pd.read_csv(path)
        total_df = total_df.append(month_df, ignore_index=True)
        print(total_df.value_counts().count())
 
    
```

```python
# Cleaning data
#Removing all NULL data from the farme

total_clean_df = total_df.dropna()

total_clean_df
```

```python
#Writing new database into the final CSV final

total_clean_df.to_csv("data/total_2020_cityBike.csv", index=False, header=True)
```
