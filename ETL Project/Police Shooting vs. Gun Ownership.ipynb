```python
import pandas as pd
import psycopg2 
import datetime as dt
from sqlalchemy import create_engine
```

```python
police_file = "Resources/fatal-police-shootings-data.csv"
police_df = pd.read_csv(police_file)
police_df.head()
```

```python
gun_own_file = "Resources/gunowners_data.csv"
gun_own_df = pd.read_csv(gun_own_file)
gun_own_df.head()
```

```python
new_police_df = police_df[['date', 'state', 'manner_of_death']].copy()
new_police_df.head()
```

```python
connection_string = "postgres:postgres@localhost:5432/fatal-police-shootings"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
engine.table_names()
```

```python
new_police_df.to_sql(name='fatal-police-shootings', con=engine, if_exists='append', index=False)
```

```python
gun_own_df.to_sql(name='gun-Owners', con=engine, if_exists='append', index=False)
```

```python
pd.read_sql_query('SELECT * FROM "fatal-police-shootings"', con=engine).head()
```

```python
pd.read_sql_query('SELECT * FROM "gun-Owners"', con=engine).head()
```
